# Folha de exercícios sobre matplotlib

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

### Exercício 1
Escreva o código Python para desenhar um gráfico de linhas com rótulo adequado no eixo x, eixo y e um título.
O código deve gerar um gráfico como o ilustrado de seguida:<br>


````python
# Adicione aqui a sua resposta
x = np.linspace(0, 50, 100)
y = 3*x
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, y)
ax.set_title('Gráfico de linha')
ax.set_xlabel('x - axis')
ax.set_ylabel('y - axis')
plt.show()
````

### Exercício 2
Escreva o código Python para desenhar uma linha usando valores de eixo fornecidos com rótulo adequado no eixo x, eixo y e um título.<br>
O código deve gerar um gráfico como o ilustrado de seguida:<br>


````python
# Adicione aqui a sua resposta
x = np.linspace(1, 3, 9)
y = [(2*x) if (x <= 2) else (-3*x + 10) for x in x]

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, y)
ax.set_title('Gráfico simples')
ax.set_xlabel('x - axis')
ax.set_ylabel('y - axis')
plt.show()
````

### Exercício 3
Escreva o código Python para gerar o gráfico de linhas com duas séries e com legendas adequadas para cada uma.<br>
O código deve gerar um gráfico como o ilustrado de seguida:<br>


````python
# Adicione aqui a sua resposta
x = np.linspace(10, 30, 9)
y1 = [(2*x) if (x <= 20) else (-3*x + 100) for x in x]
y2 = [(-3*x + 70) if (x <= 20) else (2*x - 30) for x in x]

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, y1, label='linha 1')
ax.plot(x, y2, label='linha 2')
ax.set_title('Gráfico com duas séries')
ax.set_xlabel('x - axis')
ax.set_ylabel('y - axis')
ax.legend()
plt.show()
````

### Exercício 4
Escreva o código Python para gerar o gráfico a seguir reproduzido.<br>
Note que a escala de ambos os eixos se situa entre 1 e 8.<br>


````python
# Adicione aqui a sua resposta

x = [1, 4, 5, 6, 7]
y = [2, 6, 3, 6, 3]

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, y, color='red', linestyle='-.', linewidth=3,
        marker='o', markerfacecolor='blue', markersize=12)
ax.set_title('Gráfico com marcadores e linha tracejada')
ax.set_xlabel('x - axis')
ax.set_ylabel('y - axis')
ax.set_xlim([1, 8])
ax.set_ylim([1, 8])
plt.show()
````

### Exercício 5
Escreva o código Python para gerar um gráfico de barras da popularidade das linguagens de programação.<br><br>
Dados de amostra:<br>
Linguagens de programação: Java, Python, PHP, JavaScript, C#, C++<br>
Popularidade: 22,2, 17,6, 8,8, 8, 7,7, 6,7<br><br>
O código deve gerar um gráfico como o ilustrado de seguida:<br>


````python
# Adicione aqui a sua resposta
data = {'Java': 22.2, 'Python': 17.6, 'PHP': 8.8,
        'JavaScript': 8, 'C#': 7.7, 'C++': 6.7}
df = pd.DataFrame.from_dict(data, orient='index')
df.columns = ['Popularidade']

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(df.index, df['Popularidade'], color='blue')
ax.set_title('Popularidade de linguagens de programação')
ax.set_xlabel('Linguagem')
ax.set_ylabel('Popularidade')
ax.grid(True, color='red')
ax.minorticks_on()
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
````

### Exercício 7
Escreva o código Python para gerar um gráfico de barras da popularidade das linguagens de programação.<br>
Adicione um rótulo de texto acima de cada barra mostrando a sua popularidade.<br><br>
Dados de amostra:<br>
Linguagens de programação: Java, Python, PHP, JavaScript, C#, C++<br>
Popularidade: 22,2, 17,6, 8,8, 8, 7,7, 6,7<br>
O código deve gerar um gráfico como o ilustrado de seguida:<br>


````python
# Adicione aqui a sua resposta
fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.bar(df.index, df['Popularidade'], color='blue')
ax.set_title('Popularidade de linguagens de programação')
ax.set_xlabel('Linguagem')
ax.set_ylabel('Popularidade')
ax.grid(True, color='red')
ax.minorticks_on()
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.05,
            round(yval, 2), ha='center', va='bottom', fontweight='bold')

plt.show()
````

### Exercício 10
Escreva o código Python para gerar um gráfico de barras com o número de clientes por estado considerando o dataset público das encomendas efetuadas no site [Olist Store](http://www.olist.com/)<br>
Os dados podem ser descarregados a partir do PAE na pasta `Conteúdos / DADOS / olist_customers_dataset.csv`.<br>

O código deve gerar um gráfico como o ilustrado de seguida:<br>


````python
# Adicione aqui a sua resposta
df = pd.read_csv('olist_customers_dataset.csv')
data = df.groupby('customer_state')[
    'customer_id'].count().sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(16, 5))
bars = ax.bar(data.index, data.values, color='blue')
ax.set_title('Encomendas por estado')
ax.set_xlabel('Estado')
ax.set_ylabel('Número de encomendas')
ax.grid(True, alpha=0.5, color='gray', linewidth=0.5)

for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.05,
            yval, ha='center', va='bottom')

plt.show()
````
