# Gráficos relacionais (numerical x numerical)

### tips | relplot | total_bill vs tip - day

```python
# Adicione aqui a sua resposta
sns.relplot(data=tips, x='total_bill', y='tip',
            col='day', col_wrap=2, color='purple')
plt.show()
```

### insurance | scatterplot | age vs charges   -   smoker

```python
# Adicione aqui a sua resposta
sns.scatterplot(data=insurance, x='age', y='charges',
                hue='smoker', palette='BuGn_r')
plt.title('Relation between age and charges')
plt.show()
```

### tips | scatterplot | total_bill vs tip - day

```python
# Adicione aqui a sua resposta
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', palette='colorblind').set_title(
    'Relation between total bill and tip')
plt.show()
```

### tips | lineplot | total_bill vs tip - day

```python
# Adicione aqui a sua resposta
sns.lineplot(data=tips, x='total_bill', y='tip',
             hue='day', palette='colorblind').set_title('Relation between total bill and tip')
plt.show()
```

### bank | scatterplot | NumOfProducts vs Age   -   Exited

```python
# Adicione aqui a sua resposta
sns.scatterplot(data=bank, x='NumOfProducts', y='Age', hue='Exited')
plt.show()
```

### stroke | scatterplot | bmi vs age   -   stroke

```python
# Adicione aqui a sua resposta
sns.scatterplot(data=stroke, x='bmi', y='age', hue='stroke', palette='flare')
plt.show()
```

### house | scatterplot | sqft_living vs price

```python
# Adicione aqui a sua resposta
sns.scatterplot(data=house, x='sqft_living', y='price', color='black')
plt.show()
```
